# Importing the necessary libraries


```python
import pandas as pd
import datetime as dt
```

# Data loading

```python
accident = pd.read_excel(r"Nov_13th_data\accident_data.xlsx")

accident
```

# Dealing with 1st Objective

```python
# Extracting the required column
dt_col = accident["DATE_TIME"].tolist()
dt_col
# converting the column to python datetime object
clean_dt = []
for x in dt_col:
    clean_dt.append(dt.datetime.strptime(x,"%Y-%m-%d %H:%M:%S"))
clean_dt
# extracting the month of accident occurence
month_accid = []
for x in clean_dt:
    month_accid.append(x.strftime("%B"))
month_accid
# attempt to get the total accident by month
uniq_mth = set(month_accid)
tot_accid_month = []
for x in uniq_mth:
    tot_accid_month.append((x, month_accid.count(x)))
tot_accid_month
season = {"Spring":["March","April","May"], 
          "Summer":["June","July","August"],
          "Winter":["December","January","Febuary"]}

accid_season = []
for x in month_accid:
    if x in season["Spring"]:
        accid_season.append("Spring")
    elif x in season["Summer"]:
        accid_season.append("Summer")
    elif x in season["Winter"]:
        accid_season.append("Winter")
    else:
        accid_season.append("Autum")

accid_season

```
